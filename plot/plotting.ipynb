# Importing Packages

```python
# Import required packages
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pylab import cm
```

# Loading Data

```python
# Use numpy.loadtxt to import our data
wavelength, samp_1_abs, samp_2_abs = np.loadtxt('xyy.csv', unpack=True, delimiter=',', skiprows=1)
```

# Plotting Our Data

```python
# Create figure and add axes object
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# Plot and show our data
ax.plot(wavelength, samp_1_abs)
ax.plot(wavelength, samp_2_abs)
plt.show()
```

### Removing Spines

# Final Edited Plot

### Plotting Script

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# Edit the font, font size, and axes width
mpl.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 28
plt.rcParams['axes.linewidth'] = 2

# Use numpy.loadtxt to import our data
wavelength, samp_1_abs, samp_2_abs = np.loadtxt('xyy.csv', unpack=True, delimiter=',', skiprows=1)

# Create figure object and store it in a variable called 'fig'
fig = plt.figure()

# Add axes object to our figure that takes up entire figure
ax = fig.add_axes([0, 0, 1, 1])

# Add the x and y-axis labels
ax.set_xlabel(r'$\mathregular{\lambda}$ (nm)', labelpad=10)
ax.set_ylabel('Absorbance', labelpad=10)

# Edit the major and minor ticks of the x and y axes
ax.xaxis.set_tick_params(which='major', size=12, width=2, direction='out')
ax.xaxis.set_tick_params(which='minor', size=6, width=2, direction='out')
ax.yaxis.set_tick_params(which='major', size=12, width=2, direction='in', right='on')
ax.yaxis.set_tick_params(which='minor', size=6, width=2, direction='in', right='on')

# Edit the major and minor tick locations of x and y axes
ax.xaxis.set_major_locator(mpl.ticker.MultipleLocator(100))
ax.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(20))
ax.yaxis.set_major_locator(mpl.ticker.MultipleLocator(0.5))
ax.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.1))

# Set the axis limits
ax.set_xlim(340, 940)
ax.set_ylim(0, 2.0)

axT = ax.secondary_xaxis('top')
axT.xaxis.set_tick_params(which='major', size=12, width=2, direction='in')
axT.xaxis.set_tick_params(which='minor', size=6, width=2, direction='in')
axT.xaxis.set_ticklabels([])
axT.set_xlim(340, 940)
axT.xaxis.set_major_locator(mpl.ticker.MultipleLocator(100))
axT.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(20))

# Plot the two sample absorbances, using previously generated colors
ax.plot(wavelength, samp_1_abs, linewidth=2, 
        color='blue', linestyle='solid')
ax.plot(wavelength, samp_2_abs, linewidth=2, 
        color='green', linestyle='dashed')

ax.annotate('Sample #1', color='blue',
            xy=(530, 1.45), xycoords='data',
            xytext=(850, 1.8), textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3",
                            color='blue'),
            horizontalalignment='right', verticalalignment='top')

ax.annotate('Sample #2', color='green',
            xy=(570, 1.2), xycoords='data',
            xytext=(850, 1.5), textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3",
                            color='green'),
            horizontalalignment='right', verticalalignment='top')

# Save figure
plt.savefig('mpl2ndplot.pdf', dpi=600, transparent=False, bbox_inches='tight')

# Show figure
plt.show()
```
