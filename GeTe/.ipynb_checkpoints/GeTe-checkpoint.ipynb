```python
# Import required packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# Edit the font, font size, and axes width
mpl.rcParams['font.family'] = 'Sans'
plt.rcParams['font.size'] = 16
plt.rcParams['axes.linewidth'] = 1.5

# Create figure object
fig = plt.figure(figsize=(6, 3))

# Add x and y-axes 
axE = fig.add_axes([0.00, 0.0, 0.60, 1])
axD = fig.add_axes([0.70, 0.0, 0.30, 1])

# Add x and y-axis labels 
axE.set_xlabel('High-symmetry points')
axE.set_ylabel('Energy (eV)')
axD.set_xlabel('DOS (states/eV/u.c.)')

# Hide x-axis original tick marks for energy
axE.xaxis.set_tick_params(which='both', 
                          bottom=False,
                          top=False)

# Major tick interval in DOS x-axis
axD.xaxis.set_major_locator(
    mpl.ticker.MultipleLocator(1))
# Minor tick interval in DOS x-axis
axD.xaxis.set_minor_locator(
    mpl.ticker.MultipleLocator(0.2))

# Major tick interval in y-axis every 1 eV
axE.yaxis.set_major_locator(
    mpl.ticker.MultipleLocator(1))
axD.yaxis.set_major_locator(
    mpl.ticker.MultipleLocator(1))

# Minor tick interval in y-axis every 0.2 eV
axE.yaxis.set_minor_locator(
    mpl.ticker.MultipleLocator(0.2))
axD.yaxis.set_minor_locator(
    mpl.ticker.MultipleLocator(0.2))

# Edit major and minor ticks of x and y axes
axE.yaxis.set_tick_params(which='major',
                          size=7.5, 
                          direction='in',
                          right='on')
axE.yaxis.set_tick_params(which='minor',
                          size=4.5,
                          direction='in',
                          right='on')
axD.xaxis.set_tick_params(which='major',
                          size=7.5,
                          direction='in',
                          right='on')
axD.xaxis.set_tick_params(which='minor',
                          size=4.5,
                          direction='in',
                          right='on')
axD.yaxis.set_tick_params(which='major',
                          size=7.5,
                          direction='in',
                          right='on')
axD.yaxis.set_tick_params(which='minor',
                          size=4.5,
                          direction='in',
                          right='on')

#-------------------------------------------#
prefix = 'GeTe'
#-------------------------------------------#

# Fermi energy from ./out/*.xml :
HOMO  = 2.635710373417968e-1
LUMO  = 2.758546355831248e-1
unitE = 27.2114
E_F   = (HOMO+LUMO)*unitE/2

# Minimum and maximum energies to be plotted
Emin = -2.0
Emax =  2.0

# Open and read DOS data from *.dos file
edos,dos,intdos = np.loadtxt(prefix+'.dos',
                             unpack=True,
                             skiprows=1)

# Number of bands and k-points from *.bands
nbnd = 30; nks = 201

# Open *.bands.gnu file
fbnd = open(prefix+'.bands.gnu', 'r')
f_bnd = [line for line 
         in fbnd.readlines() if line.strip()]
fbnd.close()

# Read k-points
rks = []
for ik in range(nks):
    rks.append(float(f_bnd[ik].split()[0]))

# Set high-symmetry points from *.nscf.in file
rW = rks[0]  ; rL = rks[50] ; rU = rks[80]
rX = rks[100]; rG = rks[150]; rK = rks[200]

# Read band index at each k-point
band = []
for ib in range (nbnd):
    band.append([])
    for ik in range(nks):
        tmp = f_bnd[ib*nks+ik].split()[1]
        band[ib].append(float(tmp)-E_F)
        
# Read eigen-energy at each band index
ener = []
for ik in range(nks):
    ener.append([])
    for ib in range(nbnd):
        tmp = band[ib][ik]
        ener[ik].append(float(tmp))
        
# Limit + tick labels of high-symmetry points
axE.set_xlim(rW, rK)
axE.set_xticks([rW, rL, rU, rX, rG, rK])
axE.set_xticklabels(['W', 'L', 'U', 
                     'X', '$\Gamma$', 'K'])
axE.set_ylim(Emin,Emax)
axD.set_xlim(0,2)
axD.set_ylim(Emin,Emax)

# Add dotted lines for high-symmetry points
axE.axvline(rW, color='grey', linewidth=1)
axE.axvline(rL, color='grey', linewidth=1)
axE.axvline(rU, color='grey', linewidth=1)
axE.axvline(rX, color='grey', linewidth=1)
axE.axvline(rG, color='grey', linewidth=1)
axE.axvline(rK, color='grey', linewidth=1)

# Add dotted line at Fermi energy
axE.axhline(0, color= 'grey', 
            linestyle=':', linewidth=1)
axD.axhline(0, color= 'grey',
            linestyle=':', linewidth=1)

# Plot band structure and DOS
axE.plot(rks, ener, color='blue')
axD.plot(dos, edos-E_F, color='red')

# Save figure
plt.savefig('GeTeEnDOS.pdf',bbox_inches='tight')
```
